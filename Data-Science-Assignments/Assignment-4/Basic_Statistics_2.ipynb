```python
import numpy as np
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
```

```python
df1=pd.read_csv('/content/Patient_Data.csv')
df2=pd.read_csv('/content/Billing_Data.csv')
```

```python
df1.info()
```

```python
column= ['PatientID', 'Department', 'Doctor', 'BillAmount']
df_bill=df1[column]
df_bill
```

```python
df1.drop(columns=['ReceptionistID', 'CheckInTime'], inplace=True)
df1
```

```python
dept_bill=df1.groupby('Department')['BillAmount'].sum()
dept_bill
```

```python
df1_dup=df1.drop_duplicates(subset='PatientID', keep='first')
```

```python
df1_dup
```

```python
mean=df1_dup['BillAmount'].mean()
df1_dup['BillAmount'].fillna(mean, inplace=True)
df1_dup
```

```python
df_merge=pd.merge(df1_dup, df2, on='PatientID',how='left')
df_merge
```

```python
new_patients = pd.DataFrame({
    'PatientID': [106, 107],
    'Name': ['Ravi', 'Meena'],
    'Department': ['Dermatology', 'Oncology'],
    'Doctor': ['Dr. Singh', 'Dr. Kapoor'],
    'BillAmount': [4500, 7500],
    'BillDate': ['2025-10-06', '2025-10-07'],
    'PaymentMode': ['Cash', 'Insurance']
})
```

```python
df_newpatient=pd.concat([df_merge, new_patients], ignore_index=True)
df_newpatient
```

```python
new_columns= pd.DataFrame({
    'InsuranceCovered': [True, False, True, False, True, False, True],
    'FinalAmount': [4800, 7000, 6200, 4000, 5800, 4500, 7500]})
```

```python
df_final=pd.concat([df_newpatient, new_columns], axis=1)
df_final
```
